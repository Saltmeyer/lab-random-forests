```python
import pandas as pd # consider using the up to date files in the folder
import numpy as np
pd.set_option('display.max_columns', None)
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
warnings.filterwarnings('ignore')

numerical = pd.read_csv('numerical.csv')
categorical = pd.read_csv('categorical.csv')
targets = pd.read_csv('target.csv')
data = pd.concat([numerical, categorical, targets], axis = 1)
data['TARGET_B'].value_counts()
```

## Part II Final regression model

```python
# Create new dataframe with all those who donated
donors = data[data['TARGET_B'] == 1]
```

```python
donors = donors.drop(['TARGET_B'], axis = 1)
```

## Feature Selection

```python
print(donors.shape)
donors.head()
```

```python
# Splitting into numerical and categorical
numerical = donors.select_dtypes(np.number)
categorical = donors.select_dtypes(np.object)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
numerical_scaled = scaler.fit_transform(numerical)
```

```python
column_list = pd.Series.tolist(numerical.columns)
```

```python
numerical_scaled = pd.DataFrame(data = numerical_scaled, columns= column_list)
```

```python
X = numerical_scaled.drop(['TARGET_D'], axis=1)
y = numerical_scaled['TARGET_D']
```

```python
from sklearn.feature_selection import RFE
from sklearn import linear_model
lm = linear_model.LinearRegression()
rfe = RFE(lm, n_features_to_select=30, verbose=False)
rfe.fit(X, y)
```

```python
df = pd.DataFrame(data = rfe.ranking_, columns=['Rank'])
```

```python
df['Column_name'] = pd.DataFrame(numerical.columns)

features_select = df[df['Rank']==1]
```

## Models

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_score 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
```

```python
col_names = X.columns.tolist()
```

```python
#col_names
```

```python
model = DecisionTreeRegressor()

# need to randomize X and y, but together
donor_df = pd.concat([pd.DataFrame(X), pd.Series(y)], axis=1)
```

```python
X_rand=np.array(donor_df.loc[:, donor_df.columns != 'TARGET_D']) # x_rand is where all columns are unequal to y
y_rand=np.array(donor_df['TARGET_D'])

scores = cross_val_score(model, X_rand, y_rand, cv=10)
scores
```

```python
model1 = DecisionTreeRegressor()
model2 = LinearRegression()
model3 = KNeighborsRegressor()
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.25)
```

```python
model_pipeline = [model1, model2, model3]
model_names = ['Decision Tree Regressor', 'Linear Regression', 'KNN']
scores = {}
for model, model_name in zip(model_pipeline, model_names): # review zip function
    mean_score = np.mean(cross_val_score(model, X_train, y_train, cv=10))
    scores[model_name] = mean_score
print(scores)
```
